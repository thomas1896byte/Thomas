## Solution


````python
import numpy as np

def compute_rubrene_bands(Nk_alpha, Nk_beta):
    '''Evaluate the four-band HOMO dispersion of rubrene on a 2D k-grid.

    Parameters:
      Nk_alpha : int
          Number of k-grid points along alpha (b-axis) direction.
      Nk_beta : int
          Number of k-grid points along beta (c-axis) direction.

    Returns:
      bands : np.ndarray, shape (Nk_alpha, Nk_beta, 4)
          Band energies in meV, sorted ascending along the last axis
          (band 0 = lowest energy) at each k-point.
      dk_alpha : float
          k-grid spacing along alpha in Angstrom^{-1}.
      dk_beta : float
          k-grid spacing along beta in Angstrom^{-1}.
    '''
    # Validate inputs
    if not isinstance(Nk_alpha, int) or not isinstance(Nk_beta, int):
        raise TypeError("Nk_alpha and Nk_beta must be integers")

    if Nk_alpha <= 0 or Nk_beta <= 0:
        raise ValueError("Nk_alpha and Nk_beta must be positive")

    # b=7.170 A -> alpha direction, c=14.211 A -> beta direction
    b = 7.170
    c = 14.211

    # Polaron renormalization applied to ALL transfer integrals
    P_tilde  = 0.72
    eps_AAb  = P_tilde * 134.0
    eps_AA2b = P_tilde * (-10.7)
    eps_AC   = P_tilde * 28.6
    eps_AB   = P_tilde * 4.1

    dk_alpha = 2 * np.pi / (b * Nk_alpha)
    dk_beta  = 2 * np.pi / (c * Nk_beta)

    # Half-open uniform k-grid
    ka = np.linspace(-np.pi / b, np.pi / b, Nk_alpha, endpoint=False)
    kb = np.linspace(-np.pi / c, np.pi / c, Nk_beta,  endpoint=False)

    KA, KB = np.meshgrid(ka, kb, indexing='ij')

    # Common dispersion independent of s_AC, s_AB
    common  = (2 * eps_AAb  * np.cos(KA * b)
             + 2 * eps_AA2b * np.cos(2 * KA * b))

    # AC coupling at k_gamma=0: cos(k_gamma*a/2)=1
    term_AC = 4 * eps_AC * np.cos(KA * b / 2) * np.cos(KB * c / 2)

    # AB coupling at k_gamma=0: cos(k_gamma*a/2)=1
    term_AB = 4 * eps_AB * np.cos(KA * b / 2)

    # Four bands from independent sign choices (s_AC, s_AB) in {+1,-1} x {+1,-1}
    e_pp = common + term_AC + term_AB
    e_pm = common + term_AC - term_AB
    e_mp = common - term_AC + term_AB
    e_mm = common - term_AC - term_AB

    bands_unsorted = np.stack([e_pp, e_pm, e_mp, e_mm], axis=-1)
    bands = np.sort(bands_unsorted, axis=-1)

    return bands, dk_alpha, dk_beta
````

## Solution


````python
import numpy as np
from scipy import ndimage


def compute_rubrene_bands(Nk_alpha, Nk_beta):
    '''Evaluate the four-band HOMO dispersion of rubrene on a 2D k-grid.'''
    b = 7.170
    c = 14.211
    P_tilde  = 0.72
    eps_AAb  = P_tilde * 134.0
    eps_AA2b = P_tilde * (-10.7)
    eps_AC   = P_tilde * 28.6
    eps_AB   = P_tilde * 4.1
    dk_alpha = 2 * np.pi / (b * Nk_alpha)
    dk_beta  = 2 * np.pi / (c * Nk_beta)
    ka = np.linspace(-np.pi / b, np.pi / b, Nk_alpha, endpoint=False)
    kb = np.linspace(-np.pi / c, np.pi / c, Nk_beta,  endpoint=False)
    KA, KB = np.meshgrid(ka, kb, indexing='ij')
    common  = (2 * eps_AAb  * np.cos(KA * b)
             + 2 * eps_AA2b * np.cos(2 * KA * b))
    term_AC = 4 * eps_AC * np.cos(KA * b / 2) * np.cos(KB * c / 2)
    term_AB = 4 * eps_AB * np.cos(KA * b / 2)
    e_pp = common + term_AC + term_AB
    e_pm = common + term_AC - term_AB
    e_mp = common - term_AC + term_AB
    e_mm = common - term_AC - term_AB
    bands_unsorted = np.stack([e_pp, e_pm, e_mp, e_mm], axis=-1)
    bands = np.sort(bands_unsorted, axis=-1)
    return bands, dk_alpha, dk_beta


def classify_orbits_and_areas(bands, dk_alpha, dk_beta, EF):
    '''Classify Fermi-surface orbits and compute orbit areas at a single Fermi energy.

    Parameters:
      bands : np.ndarray, shape (Nk_alpha, Nk_beta, 4)
          Band energies in meV (sorted ascending along last axis).
          Typically obtained from compute_rubrene_bands(Nk_alpha, Nk_beta).
      dk_alpha : float
          k-grid spacing along alpha in Angstrom^{-1}.
      dk_beta : float
          k-grid spacing along beta in Angstrom^{-1}.
      EF : float
          Fermi energy in meV.

    Returns:
      Ae : float
          Total electron-like orbit area in Angstrom^{-2}.
      Ah : float
          Total hole-like orbit area in Angstrom^{-2}.
    '''
    Nka, Nkb, nbands = bands.shape
    cell_area = dk_alpha * dk_beta

    Ae_total = 0.0
    Ah_total = 0.0

    # 4-connectivity (nearest neighbour only), as required by the prompt.
    structure = ndimage.generate_binary_structure(2, 1)

    for band_idx in range(nbands):
        band = bands[:, :, band_idx]

        occ   = band <= EF
        unocc = ~occ

        # Electron-like orbits: compact OCCUPIED regions not wrapping the periodic BZ.
        # Tile 3x3 so connected components that wrap the BZ stay connected.
        if np.any(occ) and not np.all(occ):
            occ_tiled = np.tile(occ, (3, 3))
            labeled_occ, _ = ndimage.label(occ_tiled, structure=structure)

            center_labels_e = set(labeled_occ[Nka:2*Nka, Nkb:2*Nkb].ravel())
            center_labels_e.discard(0)

            for lbl in center_labels_e:
                component_mask = (labeled_occ == lbl)
                coords = np.argwhere(component_mask[Nka:2*Nka, Nkb:2*Nkb])
                if coords.size == 0:
                    continue

                x_span = coords[:, 0].max() - coords[:, 0].min()
                y_span = coords[:, 1].max() - coords[:, 1].min()

                wraps = (x_span > 0.8 * Nka) or (y_span > 0.8 * Nkb)

                if not wraps:
                    count = np.sum(component_mask[Nka:2*Nka, Nkb:2*Nkb])
                    Ae_total += count * cell_area

        # Hole-like orbits: compact UNOCCUPIED regions not wrapping the periodic BZ.
        if np.any(unocc) and not np.all(unocc):
            unocc_tiled = np.tile(unocc, (3, 3))
            labeled_unocc, _ = ndimage.label(unocc_tiled, structure=structure)

            center_labels_h = set(labeled_unocc[Nka:2*Nka, Nkb:2*Nkb].ravel())
            center_labels_h.discard(0)

            for lbl in center_labels_h:
                component_mask = (labeled_unocc == lbl)
                coords = np.argwhere(component_mask[Nka:2*Nka, Nkb:2*Nkb])
                if coords.size == 0:
                    continue

                x_span = coords[:, 0].max() - coords[:, 0].min()
                y_span = coords[:, 1].max() - coords[:, 1].min()

                wraps = (x_span > 0.8 * Nka) or (y_span > 0.8 * Nkb)

                if not wraps:
                    count = np.sum(component_mask[Nka:2*Nka, Nkb:2*Nkb])
                    Ah_total += count * cell_area

    return Ae_total, Ah_total
````

# Subproblem 3


## Prompt


Compute the dimensionless semi-classical Hall conductivity profile
$\sigma_H^{\text{norm}}(E_F) = -[A_e(E_F) - A_h(E_F)] / \Omega_{BZ}$
over an array of Fermi energies spanning the HOMO bandwidth at $T = 0$.

This quantity arises from the $T = 0$ reduction of the semi-classical
Kubo-like formula:
$$\sigma_H^{sc} = \frac{e}{B_\gamma\, b c} \int dE\, f(E)\,
\frac{d}{dE}\frac{A_e(E) - A_h(E)}{\Omega_{BZ}}$$

At $T = 0$, $f(E) = \Theta(E_F - E)$, and integration by parts yields:
$$\sigma_H^{sc}(E_F) = \frac{e}{B_\gamma\, b c}
\frac{A_e(E_F) - A_h(E_F)}{\Omega_{BZ}}$$

with boundary terms vanishing at the band edges.

The sign convention is as follows: near the band top (low hole doping),
$A_h > 0$ and $A_e = 0$, so
$\sigma_H^{\text{norm}} = +A_h / \Omega_{BZ} > 0$,
consistent with a positive Hall coefficient for hole carriers
($q = +|e|$).

The Brillouin zone area is $\Omega_{BZ} = (2\pi/b)(2\pi/c)$ with
$b = 7.170$ Å and $c = 14.211$ Å.

The implementation must be fully self-contained. For each $E_F$ in
the input array, compute the four-band rubrene dispersion directly
from the tight-binding model. The band energies at wavevector
$(k_\alpha, k_\beta)$ with $k_\gamma = 0$ are indexed by two
independent sign labels $s_{AC}, s_{AB} \in \{+1, -1\}$:

$$\varepsilon(k_\alpha, k_\beta, s_{AC}, s_{AB}) =
2\tilde\varepsilon_{AA+b}\cos(k_\alpha b) +
2\tilde\varepsilon_{AA+2b}\cos(2k_\alpha b) +
s_{AC}\,4\tilde\varepsilon_{AC}
\cos\!\left(\tfrac{k_\alpha b}{2}\right)
\cos\!\left(\tfrac{k_\beta c}{2}\right) +
s_{AB}\,4\tilde\varepsilon_{AB}
\cos\!\left(\tfrac{k_\alpha b}{2}\right)$$

where all transfer integrals carry polaron renormalization
$\tilde\varepsilon_X = \tilde{P} \cdot \varepsilon_X$ with
$\tilde{P} = 0.72$. Bare transfer integrals:
$\varepsilon_{AA+b} = 134.0$ meV,
$\varepsilon_{AA+2b} = -10.7$ meV,
$\varepsilon_{AC} = 28.6$ meV,
$\varepsilon_{AB} = 4.1$ meV.
The $\alpha$-direction corresponds to lattice constant $b = 7.170$ Å
and the $\beta$-direction to $c = 14.211$ Å. The $k$-grid is uniform
with $N_{k\alpha}$ points spanning $k_\alpha \in [-\pi/b, \pi/b)$
and $N_{k\beta}$ points spanning $k_\beta \in [-\pi/c, \pi/c)$
(endpoint excluded). At each grid point sort the four energies
ascending (band 0 = lowest).

Then, for each $E_F$, classify Fermi-surface orbits directly from
the $k$-space occupancy map:

- **Electron-like orbit**: a connected region of occupied states
  ($\varepsilon \leq E_F$) in a band whose bounding box within the
  Brillouin zone spans no more than 80% of the grid points in either
  direction. Its $k$-space area (cell count
  $\times\, dk_\alpha \cdot dk_\beta$) contributes to $A_e$.
- **Hole-like orbit**: a connected region of unoccupied states
  ($\varepsilon > E_F$) in a band whose bounding box within the
  Brillouin zone spans no more than 80% of the grid points in either
  direction. Its area contributes to $A_h$.
- **Open orbit**: any connected region whose bounding box spans more
  than 80% of the grid points in either the $\alpha$ or $\beta$
  direction ($x\_\text{span} > 0.8 \times N_{k\alpha}$ or
  $y\_\text{span} > 0.8 \times N_{k\beta}$) contributes zero to
  both $A_e$ and $A_h$.

Connected-component analysis must use 4-connectivity
(nearest-neighbor only: up, down, left, right) and must correctly
handle the periodic boundary conditions of the Brillouin zone,
treating it as a torus. Occupied or unoccupied cells that wrap
across opposite edges of the $k$-grid belong to the same connected
component. Sum electron-like and hole-like areas separately over
all four bands, then compute:

$$\sigma_H^{\text{norm}}(E_F) =
-\frac{A_e(E_F) - A_h(E_F)}{\Omega_{BZ}}$$

The Hall-conductivity profile exhibits characteristic behavior:
$\sigma_H^{\text{norm}} \approx +1$ per fully occupied band deep
below the band top, decreasing as $E_F$ rises through the bandwidth,
with sign changes near Lifshitz-transition energies where orbit
topology changes between hole-like, open, and electron-like character.

Implement the following function:

```python
def compute_sigma_H_profile(Nk_alpha, Nk_beta, EF_array):
    '''Compute the normalized Hall conductivity profile over an array
    of Fermi energies.

    Parameters:
      Nk_alpha : int
          Number of k-grid points along alpha (b-axis) direction.
      Nk_beta : int
          Number of k-grid points along beta (c-axis) direction.
      EF_array : np.ndarray, shape (N_EF,)
          Fermi energies in meV.

    Returns:
      sigma_H_norm : np.ndarray, shape (N_EF,)
          Dimensionless normalized Hall conductivity:
          -[Ae(EF) - Ah(EF)] / Omega_BZ for each EF.
          Positive for pure hole transport near band top.
          Returns shape (0,) if EF_array is empty.
    '''
    return sigma_H_norm
```

## Background


The semi-classical Hall conductivity encodes how the Fermi-surface topology evolves across the bandwidth. At energies where all orbits are hole-like, $\sigma_H > 0$; at Lifshitz transitions the orbit character changes and $\sigma_H$ can change sign. The $T=0$ formula reduces the energy integral to an algebraic expression in the orbit areas at $E_F$, but the orbit areas themselves require the full numerical classification pipeline at each energy. This subproblem calls `compute_rubrene_bands` to obtain the band structure and then sweeps `classify_orbits_and_areas` over many Fermi energies to assemble the complete transport profile.

## Solution


````python
import numpy as np
from scipy import ndimage


def compute_sigma_H_profile(Nk_alpha, Nk_beta, EF_array):
    '''Compute the normalized Hall conductivity profile over an array
    of Fermi energies.

    Parameters:
      Nk_alpha : int
          Number of k-grid points along alpha (b-axis) direction.
      Nk_beta : int
          Number of k-grid points along beta (c-axis) direction.
      EF_array : np.ndarray, shape (N_EF,)
          Fermi energies in meV.

    Returns:
      sigma_H_norm : np.ndarray, shape (N_EF,)
          Dimensionless normalized Hall conductivity:
          -[Ae(EF) - Ah(EF)] / Omega_BZ for each EF.
          Positive for pure hole transport near band top.
          Returns shape (0,) if EF_array is empty.
    '''
    # Lattice constants: alpha -> b, beta -> c
    b = 7.170
    c = 14.211

    Omega_BZ = (2 * np.pi / b) * (2 * np.pi / c)

    # Polaron-renormalized transfer integrals
    P_tilde  = 0.72
    eps_AAb  = P_tilde * 134.0
    eps_AA2b = P_tilde * (-10.7)
    eps_AC   = P_tilde * 28.6
    eps_AB   = P_tilde * 4.1

    # k-grid spacings
    dk_alpha = 2 * np.pi / (b * Nk_alpha)
    dk_beta  = 2 * np.pi / (c * Nk_beta)

    # Half-open uniform k-grid
    ka = np.linspace(-np.pi / b, np.pi / b, Nk_alpha, endpoint=False)
    kb = np.linspace(-np.pi / c, np.pi / c, Nk_beta,  endpoint=False)
    KA, KB = np.meshgrid(ka, kb, indexing='ij')

    # Four-band tight-binding dispersion (inlined)
    common  = (2 * eps_AAb  * np.cos(KA * b)
             + 2 * eps_AA2b * np.cos(2 * KA * b))
    term_AC = 4 * eps_AC * np.cos(KA * b / 2) * np.cos(KB * c / 2)
    term_AB = 4 * eps_AB * np.cos(KA * b / 2)

    e_pp = common + term_AC + term_AB
    e_pm = common + term_AC - term_AB
    e_mp = common - term_AC + term_AB
    e_mm = common - term_AC - term_AB

    bands = np.sort(
        np.stack([e_pp, e_pm, e_mp, e_mm], axis=-1),
        axis=-1
    )

    Nka, Nkb, nbands = bands.shape
    cell_area = dk_alpha * dk_beta
    structure = ndimage.generate_binary_structure(2, 1)  # 4-connectivity

    sigma_H_norm = np.zeros(len(EF_array), dtype=float)

    for i, EF in enumerate(EF_array):

        Ae_total = 0.0
        Ah_total = 0.0

        for band_idx in range(nbands):

            band = bands[:, :, band_idx]
            occ   = band <= EF
            unocc = ~occ

            # Electron-like: compact occupied pockets not wrapping BZ
            if np.any(occ) and not np.all(occ):

                occ_tiled = np.tile(occ, (3, 3))
                labeled_occ, _ = ndimage.label(
                    occ_tiled,
                    structure=structure
                )
                center_labels = set(
                    labeled_occ[Nka:2*Nka, Nkb:2*Nkb].ravel()
                )
                center_labels.discard(0)

                for lbl in center_labels:
                    component_mask = (labeled_occ == lbl)
                    coords = np.argwhere(
                        component_mask[Nka:2*Nka, Nkb:2*Nkb]
                    )
                    if coords.size == 0:
                        continue
                    x_span = coords[:, 0].max() - coords[:, 0].min()
                    y_span = coords[:, 1].max() - coords[:, 1].min()
                    if not (x_span > 0.8 * Nka or y_span > 0.8 * Nkb):
                        Ae_total += (
                            np.sum(component_mask[Nka:2*Nka, Nkb:2*Nkb])
                            * cell_area
                        )

            # Hole-like: compact unoccupied pockets not wrapping BZ
            if np.any(unocc) and not np.all(unocc):

                unocc_tiled = np.tile(unocc, (3, 3))
                labeled_unocc, _ = ndimage.label(
                    unocc_tiled,
                    structure=structure
                )
                center_labels_h = set(
                    labeled_unocc[Nka:2*Nka, Nkb:2*Nkb].ravel()
                )
                center_labels_h.discard(0)

                for lbl in center_labels_h:
                    component_mask = (labeled_unocc == lbl)
                    coords = np.argwhere(
                        component_mask[Nka:2*Nka, Nkb:2*Nkb]
                    )
                    if coords.size == 0:
                        continue
                    x_span = coords[:, 0].max() - coords[:, 0].min()
                    y_span = coords[:, 1].max() - coords[:, 1].min()
                    if not (x_span > 0.8 * Nka or y_span > 0.8 * Nkb):
                        Ah_total += (
                            np.sum(component_mask[Nka:2*Nka, Nkb:2*Nkb])
                            * cell_area
                        )

        sigma_H_norm[i] = -(Ae_total - Ah_total) / Omega_BZ

    return sigma_H_norm
````

## Prompt


Compute the semi-classical Hall conductivity $\sigma_H^{sc}(E_F)$ of a
rubrene organic semiconductor as a function of Fermi energy $E_F$ at
$T=0$, using the four-band HOMO tight-binding dispersion on the 2D
herringbone ($\alpha\beta$) plane.

The rubrene unit cell contains four molecules (A, B, C, D) in a
herringbone arrangement with orthorhombic lattice constants
$a = 26.789$ Å ($\gamma$-direction, perpendicular to plane),
$b = 7.170$ Å ($\alpha$-direction, high-mobility axis), and
$c = 14.211$ Å ($\beta$-direction, transverse). The
crystallographic-to-transport mapping is:
$\alpha \leftrightarrow b$, $\beta \leftrightarrow c$,
$\gamma \leftrightarrow a$.

The four HOMO band energies at wavevector $\mathbf{k}$ with
$k_\gamma = 0$ are:
$$\varepsilon(\mathbf{k}, s_{AC}, s_{AB}) =
2\tilde{\varepsilon}_{AA{+}b}\cos(k_\alpha b) +
2\tilde{\varepsilon}_{AA{+}2b}\cos(2k_\alpha b) +
s_{AC}\,4\tilde{\varepsilon}_{AC}
\cos\!\left(\tfrac{k_\alpha b}{2}\right)
\cos\!\left(\tfrac{k_\beta c}{2}\right) +
s_{AB}\,4\tilde{\varepsilon}_{AB}
\cos\!\left(\tfrac{k_\alpha b}{2}\right)$$
where $s_{AC}, s_{AB} \in \{+1, -1\}$ independently label the four
bands, and all transfer integrals are polaron-renormalized:
$\tilde{\varepsilon}_X = \tilde{P} \cdot \varepsilon_X$.

Bare transfer integrals: $\varepsilon_{AA{+}b} = 134.0$ meV,
$\varepsilon_{AA{+}2b} = -10.7$ meV, $\varepsilon_{AC} = 28.6$ meV,
$\varepsilon_{AB} = 4.1$ meV. Polaron renormalization factor:
$\tilde{P} = 0.72$.

The 2D Brillouin zone spans
$k_\alpha \in [-\pi/b,\,\pi/b)$, $k_\beta \in [-\pi/c,\,\pi/c)$
(half-open, endpoint excluded), with area
$\Omega_{BZ} = (2\pi/b)(2\pi/c)$.

At each Fermi energy $E_F$, define the normalized semi-classical
Hall response at $T=0$ as:

$$\sigma_H^{\mathrm{norm}}(E_F) =
-\frac{A_e(E_F) - A_h(E_F)}{\Omega_{BZ}}$$

where $A_e(E_F)$ is the total $k$-space area of electron-like orbits
(compact occupied regions enclosing band minima) summed over all
bands, and $A_h(E_F)$ is the total area of hole-like orbits (compact
unoccupied regions enclosing band maxima). Open orbits that wrap
across Brillouin-zone boundaries contribute zero to both $A_e$ and
$A_h$. Connected components must be evaluated using 4-connectivity
(nearest-neighbor adjacency only: up, down, left, right) on the
periodic 2D $k$-grid, treating the Brillouin zone as a torus.

The sign convention: near the band top, $A_h > 0$ and $A_e = 0$,
giving $\sigma_H^{\mathrm{norm}} > 0$ (positive Hall coefficient for
holes with charge $+|e|$).

A connected component is classified as open (contributes zero to both
$A_e$ and $A_h$) if its bounding box within the BZ exceeds
$0.8 \times N_{k\alpha}$ grid points in the $\alpha$-direction
or $0.8 \times N_{k\beta}$ grid points in the $\beta$-direction.
Otherwise it is closed: electron-like if occupied, hole-like if
unoccupied. To handle pockets that straddle BZ edges or corners,
tile the occupancy map $3\times3$, label the tiled map with
4-connectivity, then measure bounding boxes in the center tile only
($[N_{k\alpha}:2N_{k\alpha},\; N_{k\beta}:2N_{k\beta}]$).

A correct implementation must satisfy all of the following:

1. Returns $-(A_e - A_h)/\Omega_{BZ}$ with the sign exactly as written.
2. Uses $3\times3$ tiling for periodic connected-component analysis.
3. Applies the 80% span rule to classify open vs closed orbits.
4. Returns `np.ndarray` of shape `(N_EF,)`, or shape `(0,)` for empty input.
5. $\sigma_H^{\mathrm{norm}} = 0$ below all bands; $> 0$ just below band top.

Implement the following function:


## Background


In organic semiconductors like rubrene, charge transport occurs through
narrow bands formed by weak van der Waals intermolecular coupling. The
Hall effect provides a key diagnostic of the transport mechanism: in the
semi-classical (Boltzmann) picture, carriers trace closed orbits in
$k$-space under a perpendicular magnetic field $B_\gamma$, and the Hall
conductivity $\sigma_H$ is determined by the enclosed $k$-space areas of
these orbits. The herringbone crystal structure of rubrene, with four
molecules per unit cell, produces four HOMO bands whose Fermi-surface
topology undergoes Lifshitz transitions — from hole-like to open to
electron-like orbits — as the Fermi energy sweeps through the bandwidth.
All transfer integrals are uniformly reduced by a polaron renormalization
factor $\tilde{P} = 0.72$ arising from coupling to high-energy
intramolecular vibrations. Computing $\sigma_H^{sc}(E_F)$ requires dense
numerical sampling of the 2D Brillouin zone, topological classification
of Fermi-surface orbits at each energy, and careful sign tracking
consistent with the hole-transport convention $q = +|e|$.

The connected-component analysis must preserve the periodic boundary
conditions of the Brillouin zone. Because the BZ is topologically a
torus — the left edge connects to the right edge and the top edge to
the bottom — a Fermi pocket sitting near a BZ boundary or corner will
appear as disconnected fragments if periodicity is ignored. The correct
approach is to tile the occupancy grid $3\times3$ before labeling, then
read back components from the center tile. A component is open (wrapping
across the BZ) if its bounding box in the center tile exceeds 80% of the
grid in either direction; open orbits contribute zero to both $A_e$ and
$A_h$. The Hall conductivity normalization is bounded by the four-band
Brillouin-zone limit, giving a physically allowed range approximately
between $-4$ and $+4$.

## Solution


````python
import numpy as np
from scipy import ndimage


def compute_rubrene_bands(Nk_alpha, Nk_beta):
    '''Evaluate the four-band HOMO dispersion of rubrene on a 2D k-grid.'''
    b = 7.170
    c = 14.211
    P_tilde  = 0.72
    eps_AAb  = P_tilde * 134.0
    eps_AA2b = P_tilde * (-10.7)
    eps_AC   = P_tilde * 28.6
    eps_AB   = P_tilde * 4.1
    dk_alpha = 2 * np.pi / (b * Nk_alpha)
    dk_beta  = 2 * np.pi / (c * Nk_beta)
    ka = np.linspace(-np.pi / b, np.pi / b, Nk_alpha, endpoint=False)
    kb = np.linspace(-np.pi / c, np.pi / c, Nk_beta,  endpoint=False)
    KA, KB = np.meshgrid(ka, kb, indexing='ij')
    common  = (2 * eps_AAb  * np.cos(KA * b)
             + 2 * eps_AA2b * np.cos(2 * KA * b))
    term_AC = 4 * eps_AC * np.cos(KA * b / 2) * np.cos(KB * c / 2)
    term_AB = 4 * eps_AB * np.cos(KA * b / 2)
    e_pp = common + term_AC + term_AB
    e_pm = common + term_AC - term_AB
    e_mp = common - term_AC + term_AB
    e_mm = common - term_AC - term_AB
    bands_unsorted = np.stack([e_pp, e_pm, e_mp, e_mm], axis=-1)
    bands = np.sort(bands_unsorted, axis=-1)
    return bands, dk_alpha, dk_beta


def compute_hall_conductivity_profile(Nk_alpha, Nk_beta, EF_array):
    '''Compute the normalized semi-classical Hall-response profile at T=0 for rubrene.

    Parameters:
      Nk_alpha : int
          Number of k-grid points along the alpha (b) direction.
      Nk_beta : int
          Number of k-grid points along the beta (c) direction.
      EF_array : np.ndarray, shape (N_EF,)
          Array of Fermi energies in meV at which to evaluate sigma_H.

    Returns:
      sigma_H : np.ndarray, shape (N_EF,)
          Dimensionless normalized Hall-response profile: -[Ae(EF) - Ah(EF)] / Omega_BZ
          for each Fermi energy. Positive for pure hole transport near band top.
    '''
    b = 7.170
    c = 14.211
    Omega_BZ = (2 * np.pi / b) * (2 * np.pi / c)

    bands, dk_alpha, dk_beta = compute_rubrene_bands(Nk_alpha, Nk_beta)
    Nka, Nkb, nbands = bands.shape
    cell_area = dk_alpha * dk_beta
    structure = ndimage.generate_binary_structure(2, 1)

    sigma_H = np.zeros(len(EF_array), dtype=float)

    for i, EF in enumerate(EF_array):

        Ae_total = 0.0
        Ah_total = 0.0

        for band_idx in range(nbands):

            band = bands[:, :, band_idx]
            occ   = band <= EF
            unocc = ~occ

            # Electron-like: compact occupied pockets not wrapping BZ
            if np.any(occ) and not np.all(occ):
                occ_tiled = np.tile(occ, (3, 3))
                labeled_e, _ = ndimage.label(occ_tiled, structure=structure)
                center_labels = set(labeled_e[Nka:2*Nka, Nkb:2*Nkb].ravel())
                center_labels.discard(0)
                for lbl in center_labels:
                    mask = (labeled_e == lbl)
                    coords = np.argwhere(mask[Nka:2*Nka, Nkb:2*Nkb])
                    if coords.size == 0:
                        continue
                    x_span = coords[:, 0].max() - coords[:, 0].min()
                    y_span = coords[:, 1].max() - coords[:, 1].min()
                    if not (x_span > 0.8 * Nka or y_span > 0.8 * Nkb):
                        Ae_total += (
                            np.sum(mask[Nka:2*Nka, Nkb:2*Nkb]) * cell_area
                        )

            # Hole-like: compact unoccupied pockets not wrapping BZ
            if np.any(unocc) and not np.all(unocc):
                unocc_tiled = np.tile(unocc, (3, 3))
                labeled_h, _ = ndimage.label(unocc_tiled, structure=structure)
                center_labels_h = set(labeled_h[Nka:2*Nka, Nkb:2*Nkb].ravel())
                center_labels_h.discard(0)
                for lbl in center_labels_h:
                    mask = (labeled_h == lbl)
                    coords = np.argwhere(mask[Nka:2*Nka, Nkb:2*Nkb])
                    if coords.size == 0:
                        continue
                    x_span = coords[:, 0].max() - coords[:, 0].min()
                    y_span = coords[:, 1].max() - coords[:, 1].min()
                    if not (x_span > 0.8 * Nka or y_span > 0.8 * Nkb):
                        Ah_total += (
                            np.sum(mask[Nka:2*Nka, Nkb:2*Nkb]) * cell_area
                        )

        sigma_H[i] = -(Ae_total - Ah_total) / Omega_BZ

    return sigma_H
````
